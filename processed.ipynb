```python
import pandas as pd 
import numpy as np 
```

```python
#read csv and drop first columns
df1=pd.read_csv('amendment_act/amendment_act.csv').drop(columns=['Unnamed: 0'], axis=1)
df1['Reference']='Amendment Act'
df1
#df = df.drop(df.columns[0], axis=1)
```

```python
df2=pd.read_csv('principal_act/principal_act.csv').drop(columns=['Unnamed: 0'], axis=1)
df2['Reference']='Principal Act'
df2
```

```python
df3=pd.read_csv('puA/pu(A).csv').drop(columns=['Unnamed: 0'], axis=1)
df3['Reference']='List of P.U. (A)'
df3
```

```python
df4=pd.read_csv('puB/pu(B).csv').drop(columns=['Unnamed: 0'], axis=1)
df4['Reference']='List of P.U. (B)'
df4
```

```python
df=pd.concat([df1,df2,df3,df4])
df=df.reset_index(drop=True)
df['Act No']=df['Act No'].str.strip()
df
```

```python
df['Act Name']=df['Title'].str.extract(r'(.+AKTA.+\n)').replace('\n','',regex=True)#.replace('AKTA .+ - ','',regex=True)
df
```

```python
#ref=df.groupby(['Act No','Act Name']).count().reset_index()[['Act No','Act Name']].drop_duplicates('Act Name',keep='first').reset_index(drop=True)#.groupby('Act Name').count().sort_values('Act No', ascending=False)  
ref=df[['Act No','Act Name']]
ref=ref.drop_duplicates('Act Name', keep='first').reset_index(drop=True)
ref=ref.rename(columns={'Act No': 'Act No Ref.'})
ref
```

```python
#regex extract number
# More efficient regex pattern
ref['Temp']=ref['Act Name'].str.extract(r'(.\d+(?:\.\d+)? -)', expand=False).str.replace('-','').str.strip()
ref['Act No Ref.']=np.where(ref['Act No Ref.'].isnull(),ref['Temp'],ref['Act No Ref.'])
ref.drop('Temp', axis=1, inplace=True)
ref=ref[ref['Act Name'].notnull()]
ref=ref[ref['Act No Ref.'].notnull()]
ref['Act No Ref.']=ref['Act No Ref.'].str.strip()  
ref
```

```python
df=pd.merge(df, ref, on='Act Name', how='left')
df['Act No']=np.where(df['Act No'].isnull(), df['Act No Ref.'], df['Act No'])
df
```

```python
df.to_csv('legislation_combined.csv')
```
